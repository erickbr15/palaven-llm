```python
from google.colab import userdata
hface_read_token = userdata.get('hface-read-token')
```

```python
import pandas as pd

# Crear un DataFrame con las columnas especificadas
data = {
    'instruction_number': [],
    'instruction': [],
    'response': [],
    'chatcompletion': []
}

df = pd.DataFrame(data)
df.set_index('instruction_number', inplace=True)

print(df)
```

```python
def add_instruction(instruction_number, instruction, response):
    df.loc[instruction_number, 'instruction'] = instruction
    df.loc[instruction_number, 'response'] = response
```

#5. Creation of GEMMA model and tokenizer

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("google/gemma-7b-it", token=hface_read_token)
model = AutoModelForCausalLM.from_pretrained("google/gemma-7b-it", device_map="auto", token=hface_read_token)
```

```python
def get_chat_completion(instruction):
  tokenized_instruction = tokenizer(instruction, return_tensors="pt").to("cuda")
  chat_completion_result = model.generate(**tokenized_instruction, max_new_tokens=500)
  chat_completion = tokenizer.decode(chat_completion_result[0])
  return chat_completion
```

```python
instruction_1 = df.loc[1, 'instruction']
chat_completion_1 = get_chat_completion(instruction_1)
df.loc[1, 'chatcompletion'] = chat_completion_1

print('chat_completion_1 done...')

instruction_2 = df.loc[2, 'instruction']
chat_completion_2 = get_chat_completion(instruction_2)
df.loc[2, 'chatcompletion'] = chat_completion_2

print('chat_completion_2 done...')
```

```python
from bert_score import score as bert_score

references = df['response'].tolist()
candidates = df['chatcompletion'].tolist()

P, R, F1 = bert_score(candidates, references, lang='es', verbose=True)
```

```python
print(P, R, F1)
```

```python
from rouge_score import rouge_scorer

scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

rouge1_precision, rouge1_recall, rouge1_f1 = [], [], []
rouge2_precision, rouge2_recall, rouge2_f1 = [], [], []
rougeL_precision, rougeL_recall, rougeL_f1 = [], [], []

for _, row in df.iterrows():
    scores = scorer.score(row['response'], row['chatcompletion'])
    rouge1_precision.append(scores['rouge1'].precision)
    rouge1_recall.append(scores['rouge1'].recall)
    rouge1_f1.append(scores['rouge1'].fmeasure)
    rouge2_precision.append(scores['rouge2'].precision)
    rouge2_recall.append(scores['rouge2'].recall)
    rouge2_f1.append(scores['rouge2'].fmeasure)
    rougeL_precision.append(scores['rougeL'].precision)
    rougeL_recall.append(scores['rougeL'].recall)
    rougeL_f1.append(scores['rougeL'].fmeasure)
```

```python
print (rouge1_precision, rouge1_recall, rouge1_f1)
print (rouge2_precision, rouge2_recall, rouge2_f1)
print (rougeL_precision, rougeL_recall, rougeL_f1)
```
