# ROUGE Evaluation Notebook

## Purpose
This notebook retrieves and visualizes ROUGE metrics (ROUGE-1, ROUGE-2, and ROUGE-L) for different inference experiments using the google/gemma-7b-it. The experiments included in this analysis are:
- `llm_vanilla`
- `llm_rag`
- `llm_finetuned`
- `llm_finetuned_rag`

The metrics are fetched via the **Palaven API**, and the results are plotted for comparison across batches. The primary goal is to evaluate how each model performs in terms of ROUGE-1, ROUGE-2, and ROUGE-L scores across multiple batches.



### 1. Setup and Initialization

```python
from google.colab import userdata
from palaven_api_v2 import PalavenApi

evaluation_session_id = 'EB9C5839-7B20-4D7D-B3F7-17528180676D'
palaven_base_url = userdata.get('palaven-base-url')

palaven_api = PalavenApi(palaven_base_url)
```

### 2. Fetching Data

```python
def fetch_rouge_score_metrics_by_exercise(exercise_name, rouge_type):
  metrics = palaven_api.fetch_rouge_metrics(evaluation_session_id, exercise_name, rouge_type)
  return metrics
```

### 3. Data processing

```python
llm_vanilla_exercise = 'llmvanilla'
llm_rag_exercise = 'llmrag'
llm_finetuned_exercise = 'llmfinetuned'
llm_finetuned_rag_exercise = 'llmfinetunedrag'
```

### 4. Data Visualization

```python
import matplotlib.pyplot as plt

metrics_r1 = ['precision', 'recall', 'f1']
metrics_r2 = ['precision', 'recall', 'f1']
metrics_rl = ['precision', 'recall', 'f1']

models = ['llm_vanilla', 'llm_rag', 'llm_finetuned', 'llm_finetuned_rag']

precision_data_r1 = [llm_vanilla_precision_r1, llm_rag_precision_r1, llm_finetuned_precision_r1, llm_finetuned_rag_precision_r1]
recall_data_r1 = [llm_vanilla_recall_r1, llm_rag_recall_r1, llm_finetuned_recall_r1, llm_finetuned_rag_recall_r1]
f1_data_r1 = [llm_vanilla_f1_r1, llm_rag_f1_r1, llm_finetuned_f1_r1, llm_finetuned_rag_f1_r1]

precision_data_r2 = [llm_vanilla_precision_r2, llm_rag_precision_r2, llm_finetuned_precision_r2, llm_finetuned_rag_precision_r2]
recall_data_r2 = [llm_vanilla_recall_r2, llm_rag_recall_r2, llm_finetuned_recall_r2, llm_finetuned_rag_recall_r2]
f1_data_r2 = [llm_vanilla_f1_r2, llm_rag_f1_r2, llm_finetuned_f1_r2, llm_finetuned_rag_f1_r2]

precision_data_rl = [llm_vanilla_precision_rl, llm_rag_precision_rl, llm_finetuned_precision_rl, llm_finetuned_rag_precision_rl]
recall_data_rl = [llm_vanilla_recall_rl, llm_rag_recall_rl, llm_finetuned_recall_rl, llm_finetuned_rag_recall_rl]
f1_data_rl = [llm_vanilla_f1_rl, llm_rag_f1_rl, llm_finetuned_f1_rl, llm_finetuned_rag_f1_rl]

fig, ax = plt.subplots(3, 3, figsize=(15, 15))

def plot_metric(data, row, col, metric_name, title):
    for i, model in enumerate(models):
        ax[row, col].plot(data[i], label=model)
    ax[row, col].set_title(title)
    ax[row, col].set_xlabel('Batch')
    ax[row, col].set_ylabel(metric_name)
    ax[row, col].legend(loc='lower right')
    ax[row, col].grid(True, which='both', linestyle='--', linewidth=0.5, color='gray')
    ax[row, col].set_facecolor('lightgray')

# Rouge-1
plot_metric(precision_data_r1, 0, 0, 'Precision', 'ROUGE-1 Precision')
plot_metric(recall_data_r1, 1, 0, 'Recall', 'ROUGE-1 Recall')
plot_metric(f1_data_r1, 2, 0, 'F1-Score', 'ROUGE-1 F1-Score')

# Rouge-2
plot_metric(precision_data_r2, 0, 1, 'Precision', 'ROUGE-2 Precision')
plot_metric(recall_data_r2, 1, 1, 'Recall', 'ROUGE-2 Recall')
plot_metric(f1_data_r2, 2, 1, 'F1-Score', 'ROUGE-2 F1-Score')

# Rouge-L
plot_metric(precision_data_rl, 0, 2, 'Precision', 'ROUGE-L Precision')
plot_metric(recall_data_rl, 1, 2, 'Recall', 'ROUGE-L Recall')
plot_metric(f1_data_rl, 2, 2, 'F1-Score', 'ROUGE-L F1-Score')

plt.tight_layout()
plt.show()
```
