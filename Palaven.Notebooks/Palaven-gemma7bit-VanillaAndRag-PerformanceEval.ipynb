```python
from palaven_api_helper import fetch_vanilla_chat_completions, fetch_rag_chat_completions
import pandas as pd
```

```python
evaluation_session_id = '9DB25876-D92E-4CD0-9C20-7E6802387A5D'
batch_number = 1
```

```python
data_shape = {
    'instruction_id': [],
    'evaluation_session_id': [],
    'dataset_id': [],
    'batch_size': [],
    'large_language_model': [],
    'device_info': [],
    'excercise_type': [],
    'batch_number': [],
    'instruction': [],
    'response': [],
    'category': [],
    'response_to_evaluate': [],
    'elapsed_time': []
}

vanilla_df = pd.DataFrame(data_shape)
vanilla_df.set_index('instruction_id', inplace=True)

rag_df = pd.DataFrame(data_shape)
rag_df.set_index('instruction_id', inplace=True)
```

```python
def add_vanilla_response(vanilla_response):
  vanilla_df.loc[vanilla_response['instructionId'], 'evaluation_session_id'] = vanilla_response['evaluationSessionId']
  vanilla_df.loc[vanilla_response['instructionId'], 'dataset_id'] = vanilla_response['datasetId']
  vanilla_df.loc[vanilla_response['instructionId'], 'batch_size'] = vanilla_response['batchSize']
  vanilla_df.loc[vanilla_response['instructionId'], 'large_language_model'] = vanilla_response['largeLanguageModel']
  vanilla_df.loc[vanilla_response['instructionId'], 'device_info'] = vanilla_response['deviceInfo']
  vanilla_df.loc[vanilla_response['instructionId'], 'excercise_type'] = vanilla_response['chatCompletionExcerciseType']
  vanilla_df.loc[vanilla_response['instructionId'], 'batch_number'] = vanilla_response['batchNumber']
  vanilla_df.loc[vanilla_response['instructionId'], 'instruction'] = vanilla_response['instruction']
  vanilla_df.loc[vanilla_response['instructionId'], 'response'] = vanilla_response['response']
  vanilla_df.loc[vanilla_response['instructionId'], 'category'] = vanilla_response['category']
  vanilla_df.loc[vanilla_response['instructionId'], 'response_to_evaluate'] = vanilla_response['llmResponseToEvaluate']
  vanilla_df.loc[vanilla_response['instructionId'], 'elapsed_time'] = vanilla_response['elapsedTime']
```

```python
vanilla_completions = fetch_vanilla_chat_completions(evaluation_session_id, batch_number)

for item in vanilla_completions:
  add_vanilla_response(item)

print(vanilla_df)
```

```python
def add_rag_response(rag_response):
  rag_df.loc[rag_response['instructionId'], 'evaluation_session_id'] = rag_response['evaluationSessionId']
  rag_df.loc[rag_response['instructionId'], 'dataset_id'] = rag_response['datasetId']
  rag_df.loc[rag_response['instructionId'], 'batch_size'] = rag_response['batchSize']
  rag_df.loc[rag_response['instructionId'], 'large_language_model'] = rag_response['largeLanguageModel']
  rag_df.loc[rag_response['instructionId'], 'device_info'] = rag_response['deviceInfo']
  rag_df.loc[rag_response['instructionId'], 'excercise_type'] = rag_response['chatCompletionExcerciseType']
  rag_df.loc[rag_response['instructionId'], 'batch_number'] = rag_response['batchNumber']
  rag_df.loc[rag_response['instructionId'], 'instruction'] = rag_response['instruction']
  rag_df.loc[rag_response['instructionId'], 'response'] = rag_response['response']
  rag_df.loc[rag_response['instructionId'], 'category'] = rag_response['category']
  rag_df.loc[rag_response['instructionId'], 'response_to_evaluate'] = rag_response['llmResponseToEvaluate']
  rag_df.loc[rag_response['instructionId'], 'elapsed_time'] = rag_response['elapsedTime']
```

```python
rag_completions = fetch_rag_chat_completions(evaluation_session_id, batch_number)

for item in rag_completions:
  add_rag_response(item)

print(rag_df)
```

```python
from bert_score import score as bert_score

vanilla_references = vanilla_df['response'].tolist()
vanilla_candidates = vanilla_df['response_to_evaluate'].tolist()

precision_vanilla, recall_vanilla, f1_vanilla = bert_score(vanilla_candidates, vanilla_references, lang='es', verbose=True)

print(f'Precision: {precision_vanilla}')
print(f'Recall: {recall_vanilla}')
print(f'F1: {f1_vanilla}')
```

```python
rag_references = rag_df['response'].tolist()
rag_candidates = rag_df['response_to_evaluate'].tolist()

precision_rag, recall_rag, f1_rag = bert_score(rag_candidates, rag_references, lang='es', verbose=True)
```

```python
print(f'Precision: {precision_rag}')
print(f'Recall: {recall_rag}')
print(f'F1: {f1_rag}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Vanilla Precision': precision_vanilla,
    'Vanilla Recall': recall_vanilla,
    'Vanilla F1': f1_vanilla,
    'RAG Precision': precision_rag,
    'RAG Recall': recall_rag,
    'RAG F1': f1_rag
}

metrics_df = pd.DataFrame(data)

# Calcular promedios
mean_vanilla_precision = sum(precision_vanilla) / len(precision_vanilla)
mean_vanilla_recall = sum(recall_vanilla) / len(recall_vanilla)
mean_vanilla_f1 = sum(f1_vanilla) / len(f1_vanilla)

mean_rag_precision = sum(precision_rag) / len(precision_rag)
mean_rag_recall = sum(recall_rag) / len(recall_rag)
mean_rag_f1 = sum(f1_rag) / len(f1_rag)

# Plot the metrics
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

# Precision plot
ax[0].bar(['Vanilla', 'RAG'], [mean_vanilla_precision, mean_rag_precision], color=['blue', 'green'])
ax[0].set_title('Precision Comparison')
ax[0].set_ylabel('Precision')
ax[0].set_ylim(0, 1)

# Recall plot
ax[1].bar(['Vanilla', 'RAG'], [mean_vanilla_recall, mean_rag_recall], color=['blue', 'green'])
ax[1].set_title('Recall Comparison')
ax[1].set_ylabel('Recall')
ax[1].set_ylim(0, 1)

# F1 Score plot
ax[2].bar(['Vanilla', 'RAG'], [mean_vanilla_f1, mean_rag_f1], color=['blue', 'green'])
ax[2].set_title('F1 Score Comparison')
ax[2].set_ylabel('F1 Score')
ax[2].set_ylim(0, 1)

plt.show()
```

```python
from rouge_score import rouge_scorer

rouge_scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

vanilla_r1_precision, vanilla_r1_recall, vanilla_r1_f1 = [], [], []
vanilla_r2_precision, vanilla_r2_recall, vanilla_r2_f1 = [], [], []
vanilla_rL_precision, vanilla_rL_recall, vanilla_rL_f1 = [], [], []

for _, row in vanilla_df.iterrows():
    scores = rouge_scorer.score(row['response'], row['response_to_evaluate'])

    vanilla_r1_precision.append(scores['rouge1'].precision)
    vanilla_r1_recall.append(scores['rouge1'].recall)
    vanilla_r1_f1.append(scores['rouge1'].fmeasure)

    vanilla_r2_precision.append(scores['rouge2'].precision)
    vanilla_r2_recall.append(scores['rouge2'].recall)
    vanilla_r2_f1.append(scores['rouge2'].fmeasure)

    vanilla_rL_precision.append(scores['rougeL'].precision)
    vanilla_rL_recall.append(scores['rougeL'].recall)
    vanilla_rL_f1.append(scores['rougeL'].fmeasure)
```

```python
from rouge_score import rouge_scorer

rouge_scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

rag_r1_precision, rag_r1_recall, rag_r1_f1 = [], [], []
rag_r2_precision, rag_r2_recall, rag_r2_f1 = [], [], []
rag_rL_precision, rag_rL_recall, rag_rL_f1 = [], [], []

for _, row in rag_df.iterrows():
    scores = rouge_scorer.score(row['response'], row['response_to_evaluate'])

    rag_r1_precision.append(scores['rouge1'].precision)
    rag_r1_recall.append(scores['rouge1'].recall)
    rag_r1_f1.append(scores['rouge1'].fmeasure)

    rag_r2_precision.append(scores['rouge2'].precision)
    rag_r2_recall.append(scores['rouge2'].recall)
    rag_r2_f1.append(scores['rouge2'].fmeasure)

    rag_rL_precision.append(scores['rougeL'].precision)
    rag_rL_recall.append(scores['rougeL'].recall)
    rag_rL_f1.append(scores['rougeL'].fmeasure)
```

```python
vanilla_metrics_df = pd.DataFrame({
    'rouge1_precision': vanilla_r1_precision,
    'rouge1_recall': vanilla_r1_recall,
    'rouge1_f1': vanilla_r1_f1,
    'rouge2_precision': vanilla_r2_precision,
    'rouge2_recall': vanilla_r2_recall,
    'rouge2_f1': vanilla_r2_f1,
    'rougeL_precision': vanilla_rL_precision,
    'rougeL_recall': vanilla_rL_recall,
    'rougeL_f1': vanilla_rL_f1
})

rag_metrics_df = pd.DataFrame({
    'rouge1_precision': rag_r1_precision,
    'rouge1_recall': rag_r1_recall,
    'rouge1_f1': rag_r1_f1,
    'rouge2_precision': rag_r2_precision,
    'rouge2_recall': rag_r2_recall,
    'rouge2_f1': rag_r2_f1,
    'rougeL_precision': rag_rL_precision,
    'rougeL_recall': rag_rL_recall,
    'rougeL_f1': rag_rL_f1
})

# Graficar las métricas
fig, ax = plt.subplots(3, 3, figsize=(18, 18))

# ROUGE-1 Precision
ax[0, 0].plot(vanilla_metrics_df['rouge1_precision'], label='Vanilla', color='blue')
ax[0, 0].plot(rag_metrics_df['rouge1_precision'], label='RAG', color='green')
ax[0, 0].set_title('ROUGE-1 Precision')
ax[0, 0].legend()

# ROUGE-1 Recall
ax[0, 1].plot(vanilla_metrics_df['rouge1_recall'], label='Vanilla', color='blue')
ax[0, 1].plot(rag_metrics_df['rouge1_recall'], label='RAG', color='green')
ax[0, 1].set_title('ROUGE-1 Recall')
ax[0, 1].legend()

# ROUGE-1 F1
ax[0, 2].plot(vanilla_metrics_df['rouge1_f1'], label='Vanilla', color='blue')
ax[0, 2].plot(rag_metrics_df['rouge1_f1'], label='RAG', color='green')
ax[0, 2].set_title('ROUGE-1 F1')
ax[0, 2].legend()

# ROUGE-2 Precision
ax[1, 0].plot(vanilla_metrics_df['rouge2_precision'], label='Vanilla', color='blue')
ax[1, 0].plot(rag_metrics_df['rouge2_precision'], label='RAG', color='green')
ax[1, 0].set_title('ROUGE-2 Precision')
ax[1, 0].legend()

# ROUGE-2 Recall
ax[1, 1].plot(vanilla_metrics_df['rouge2_recall'], label='Vanilla', color='blue')
ax[1, 1].plot(rag_metrics_df['rouge2_recall'], label='RAG', color='green')
ax[1, 1].set_title('ROUGE-2 Recall')
ax[1, 1].legend()

# ROUGE-2 F1
ax[1, 2].plot(vanilla_metrics_df['rouge2_f1'], label='Vanilla', color='blue')
ax[1, 2].plot(rag_metrics_df['rouge2_f1'], label='RAG', color='green')
ax[1, 2].set_title('ROUGE-2 F1')
ax[1, 2].legend()

# ROUGE-L Precision
ax[2, 0].plot(vanilla_metrics_df['rougeL_precision'], label='Vanilla', color='blue')
ax[2, 0].plot(rag_metrics_df['rougeL_precision'], label='RAG', color='green')
ax[2, 0].set_title('ROUGE-L Precision')
ax[2, 0].legend()

# ROUGE-L Recall
ax[2, 1].plot(vanilla_metrics_df['rougeL_recall'], label='Vanilla', color='blue')
ax[2, 1].plot(rag_metrics_df['rougeL_recall'], label='RAG', color='green')
ax[2, 1].set_title('ROUGE-L Recall')
ax[2, 1].legend()

# ROUGE-L F1
ax[2, 2].plot(vanilla_metrics_df['rougeL_f1'], label='Vanilla', color='blue')
ax[2, 2].plot(rag_metrics_df['rougeL_f1'], label='RAG', color='green')
ax[2, 2].set_title('ROUGE-L F1')
ax[2, 2].legend()

plt.tight_layout()
plt.show()
```
