# BLEU Evaluation Notebook

## Purpose
This notebook retrieves and visualizes the BLEU score for for different inference experiments using the google/gemma-7b-it. The experiments included in this analysis are:
- `llm_vanilla`
- `llm_rag`
- `llm_finetuned`
- `llm_finetuned_rag`

The metrics are fetched via the **Palaven API**, and the results are plotted for comparison across batches. The primary goal is to evaluate how each model performs in terms of BLEU score across multiple batches.


### 1. Setup and Initialization

```python
from google.colab import userdata
from palaven_api_v2 import PalavenApi

evaluation_session_id = 'EB9C5839-7B20-4D7D-B3F7-17528180676D'
palaven_base_url = userdata.get('palaven-base-url')

palaven_api = PalavenApi(palaven_base_url)
```

### 2. Fetching Data

```python
def fetch_bleu_score_metrics_by_exercise(exercise_name):
  metrics = palaven_api.fetch_bleu_score_metrics(evaluation_session_id, exercise_name)
  return metrics
```

### 3. Data preparation

```python
llm_vanilla_exercise = 'llmvanilla'
llm_vanilla_bleu = []

llm_vanilla_metrics = fetch_bleu_score_metrics_by_exercise(llm_vanilla_exercise)

print(llm_vanilla_metrics)

for metric in llm_vanilla_metrics:
  llm_vanilla_bleu.append(metric['bleuScore'])
```

```python
llm_rag_exercise = 'llmrag'
llm_rag_bleu = []

llm_rag_metrics = fetch_bleu_score_metrics_by_exercise(llm_rag_exercise)

print(llm_rag_metrics)

for metric in llm_rag_metrics:
  llm_rag_bleu.append(metric['bleuScore'])
```

```python
llm_finetuned_exercise = 'llmfinetuned'
llm_finetuned_bleu = []

llm_finetuned_metrics = fetch_bleu_score_metrics_by_exercise(llm_finetuned_exercise)

print(llm_finetuned_metrics)

for metric in llm_finetuned_metrics:
  llm_finetuned_bleu.append(metric['bleuScore'])
```

```python
llm_finetuned_rag_exercise = 'llmfinetunedrag'
llm_finetuned_rag_bleu = []

llm_finetuned_rag_metrics = fetch_bleu_score_metrics_by_exercise(llm_finetuned_rag_exercise)

print(llm_finetuned_rag_metrics)

for metric in llm_finetuned_rag_metrics:
  llm_finetuned_rag_bleu.append(metric['bleuScore'])
```

### 4. Data Visualization

```python
import matplotlib.pyplot as plt

models = ['llm_vanilla', 'llm_rag', 'llm_finetuned', 'llm_finetuned_rag']
bleu_data = [llm_vanilla_bleu, llm_rag_bleu, llm_finetuned_bleu, llm_finetuned_rag_bleu]

fig, ax = plt.subplots(figsize=(10, 6))

ax.set_facecolor('lightgray')
ax.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray')

for i, model in enumerate(models):
    ax.plot(bleu_data[i], label=model)

ax.set_title('BLEU Score')
ax.set_xlabel('Batch')
ax.set_ylabel('Precision')
ax.legend(loc='lower right')

plt.tight_layout()
plt.show()
```
